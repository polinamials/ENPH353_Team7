```python
import cv2
import numpy as np
import keyboard
import matplotlib.pyplot as plt

imgs = np.load("/home/fizzer/ros_ws/cnn_trainer/data/ped_full_size.npy")
imgs.shape
```

```python
# top leftcorner coords
# rows, columns
r, c = 400, 540
width = 200
height = 225

# first intersection
#i=30, person is in frame, mean=111.503
#i=40, person out of frame, mean=112.855
#average is less when the person enters the frame

#second intersection
# i = 184, person in frame, mean=114.311
# i=190, person out of frame, mean = 115.364
#average is less when the person enters the frame
i=190

test_img = cv2.cvtColor(imgs[i], cv2.COLOR_BGR2RGB)
cropped_test_img = test_img[r:r+height+1, c:c+width+1]
plt.imshow(cropped_test_img)
plt.title(str(np.mean(cropped_test_img)))
```
